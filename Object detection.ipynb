```python
import cv2
```

```python
import matplotlib.pyplot as plt
import random
```

```python
config = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config)
```

```python
Labels= []
file = 'coco.names.txt'
with open(file,'rt') as f:
        Labels = f.read().rstrip('\n').split('\n')
```

```python
print(Labels)
```

```python
model.setInputSize(320,320)
model.setInputScale(0.00784313725)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
image = cv2.imread('daryan-shamkhali-tRDGs9utMUo-unsplash.jpg')
```

```python
plt.imshow(image)
```

```python
index,confidence,bbox = model.detect(image,confThreshold=0.5)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
for indexes,confidences,box in zip(index.flatten(),confidence.flatten(), bbox):
    cv2.rectangle(image, box, (255,0,0),2)
    cv2.putText(image, Labels[indexes-1],(box[0]+10,box[1]+40), font, font_scale, color=(0,255,0),thickness=3)
```
